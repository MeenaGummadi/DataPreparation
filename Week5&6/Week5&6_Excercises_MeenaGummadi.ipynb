### Chapter 7

###### Filter Out missing data

```python
#checking the count of null values
df.isnull().sum().sort_values(ascending=False)
```

```python
# dropping all the columns having null values greater than 29185
df1 = df.dropna(axis=1, thresh=len(df)-29185)
print(df1.isnull().sum().sort_values(ascending=False))
```

```python
# dropping all the rows having null values
df1 = df1.dropna()
print(df1.isnull().sum().sort_values(ascending=False))
```

### Chapter 11

###### Convert between string and date time

```python
df2=pd.read_excel("BOING-BOING-CANDY-HIERARCHY-2016-SURVEY-Responses.xlsx")
```

```python
df2['Timestamp'].dtypes
```

```python
#convert datetime to string
df2['Timestamp']=df2['Timestamp'].astype(str)
```

```python
df2['Timestamp'].dtypes
```

```python
#convert string to datetime
df2['Timestamp']=pd.to_datetime(df2['Timestamp'])
```

```python
df2['Timestamp'].dtypes
```

###### Convert timestamps to periods and back

```python
#creating time series
timeseries= df2.set_index("Timestamp")
```

```python
# converting timestamp to seconds
ts_prd=timeseries.to_period(freq='min')
ts_prd
```

```python
# converting period to timestamp
ts_prd.to_timestamp()
```
